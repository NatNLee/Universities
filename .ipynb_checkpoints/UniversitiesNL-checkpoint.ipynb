#private or public column

```python
#create private or public column test 1
#df.iloc[:,3].head(10)
#df.iloc[:,3].tail(10)
#df.Type.str.split(" ").str[-1]
```

```python
#create private or public column test 2 final

df['Private_or_Public'] = ["Private" if x ==True else "Public" for x in df.Type.str.contains('Private')]
df
```

#for profit or not for profit column

```python
#create for profit or not for profit column test 1
#df.iloc[:,3].head(30)
#df['Profit'] = ["not-for-profit" if x ==True else None for x in df.Type.str.contains('not-for-profit')]
#df
#df['Profit'] = ["not-for-profit" if x ==True else None for x in df.Type.str.contains('not-for-profit')]
#df
#df['Profit'] = ["for-profit" if x ==True else None for x in df.Type.str.contains('for-profit')]
#above doesnt work because it overwrites not for profit since not for profit contains the words for profit
```

```python
#create profit columnn test 2 final
def conditions(p):
    if 'not-for-profit' in p.Type:
        return "not-for-profit"
    elif ' for-profit' in p.Type:    
        return "for-profit"
    else:
        pass
```

```python
df['Profit'] = df.apply(conditions, axis=1)
df
```

#primarily assc column

```python
#primarily assc column

df['primarily_associates'] = [1 if x ==True else 0 for x in df.Type.str.contains('primarily associate')]
df
```
